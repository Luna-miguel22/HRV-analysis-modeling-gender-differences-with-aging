BMI

```python
bmi_stats = df.groupby(['Age_group', 'Sex'])['BMI'].mean().reset_index()

# Optional: Make it more readable by mapping the Sex codes to labels
bmi_stats['Gender'] = bmi_stats['Sex'].map({0: 'Male', 1: 'Female'})

# Reorder columns for the final report
result = bmi_stats[['Age_group', 'Gender', 'BMI']]

print(round(result,2))
```

Length

```python
length_stats = df.groupby(['Age_group', 'Sex'])['Length'].mean().reset_index()

# Optional: Make it more readable by mapping the Sex codes to labels
length_stats['Gender'] = length_stats['Sex'].map({0: 'Male', 1: 'Female'})

# Reorder columns for the final report
result = length_stats[['Age_group', 'Gender', 'Length']]

print(round(result,2))
```

Device

```python
import pandas as pd



# Group by Device and Sex
# Device: 0 = TFM, 1 = CNAP/MP150
# Sex: 0 = Male, 1 = Female
counts = df.groupby(['Device', 'Sex']).size().unstack(fill_value=0)

# Display results for TFM (Device 0)
print(f"Device TFM (0):")
print(f"  - Males: {counts.loc[0, 0]}")
print(f"  - Females: {counts.loc[0, 1]}")

# Display results for CNAP/MP150 (Device 1)
print(f"\nDevice CNAP/MP150 (1):")
print(f"  - Males: {counts.loc[1, 0]}")
print(f"  - Females: {counts.loc[1, 1]}")
```

```python
# Group by Age_group and Device
# Device: 0 = TFM, 1 = CNAP/MP150
age_device_counts = df.groupby(['Age_group', 'Device']).size().unstack(fill_value=0)

# Rename the columns for clarity
age_device_counts.columns = ['TFM (0)', 'CNAP/MP150 (1)']

print(age_device_counts)
```
